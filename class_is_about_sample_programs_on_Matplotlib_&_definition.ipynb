**Introduction to Matplotlib**

Matplotlib is one of the most popular Python libraries used for data visualization. It helps us represent data graphically so that trends, patterns, and comparisons become easier to understand.

 > Matplotlib is already available in Google Colab.











## List of Functions and Their Usage

| Function | Usage |
|----------|--------|
| `plt.plot(x,y)` | Line plot |
| `plt.bar(x,y)` | Bar chart |
| `plt.hist(data, bins)` | Histogram |
| `plt.scatter(x,y)` | Scatter plot |
| `plt.pie(values, labels)` | Pie chart |
| `plt.xlabel(), plt.ylabel()` | Set axis labels |
| `plt.title()` | Set plot title |
| `plt.legend()` | Add legend |
| `plt.grid()` | Show grid |
| `plt.subplot()` | Multiple plots in one figure |
| `plt.figure(figsize=(w,h))` | Define figure size |
| `plt.show()` | Display plot |

**Basic Plot Structure**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y)
plt.title("Simple Line Plot")
plt.legend("s")
plt.xlabel("Nums")
plt.ylabel("Values")
plt.show()
```

##Bar Chart
- Definition

Represents data using rectangular bars.

- When to Use?

When comparing different categories.

- Applications
1. Student marks comparison
2. Product sales comparison
3. Department performance

- Best Use Case

Comparing multiple categories.

```python
students = ['raju','rani','ravi','rajitha']
marks = [85,90,78,95]

plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

##Histogram
- Definition

Shows frequency distribution of numerical data.

- When to Use?

To understand data distribution.

- Applications
1. Exam score analysis
2. Salary distribution
3. Age distribution

- Best Use Case

Understanding how data is spread.

```python
ages = [18000,19000,20000,21000,22000,23000,24000,25000]

plt.hist(ages, bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

##Scatter Plot
- Definition

Represents individual data points.

- When to Use?

To find relationships between variables.

- Applications
1. Height vs Weight
2. Study Hours vs Marks
3. Experience vs Salary
- Best Use Case

Finding correlations between variables.

```python
hours = [6,5,3,4,2,1]
marks = [90,60,50,40,75,30]

plt.scatter(hours, marks)
plt.title("Study Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.show()
```

##Pie Chart
- Definition

Represents data as portions of a whole.

- When to Use?

To show percentage contributions.

- Applications
1. Market share
2. Budget allocation
3. Survey results
- Best Use Case

Showing percentages.

```python
labels = ['Python','Java','C++','JavaScript']
values = [40,25,20,15]

plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Programming Language Usage")
plt.show()
```

```python
# multiple plots on one figure
x = [1,2,3,4,5]

plt.subplot(1,2,1)
plt.plot(x,[1,4,9,16,25])
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,[1,2,3,4,5])
plt.title("Plot 2")

plt.show()
```

```python
#Change Color
plt.plot([1,2,3],[4,5,6], color='red')
plt.show()
```

```python
from numpy._core.multiarray import dot
#Change Line Style
plt.plot([1,2,3],[4,5,6], linestyle='--')
plt.show()
```

```python
#Change Marker
plt.plot([1,2,3],[4,5,6], marker='o')
plt.grid(True)
plt.show()
```

```python
#plt.figure(figsize=(w, h)) is used to set the size of the graph (figure).
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.figure(figsize=(6, 4))
plt.plot(x, y)

plt.title("Line Plot")
plt.show()
```

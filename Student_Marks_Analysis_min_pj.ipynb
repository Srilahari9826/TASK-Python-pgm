Problem : Student Marks Analysis
Create a dataset containing marks of 10 students.

students = ['A','B','C','D','E','F','G','H','I','J']

marks = [78,85,92,67,88,95,72,81,90,76] Tasks

Display student marks graphically.
Find the highest scorer.
Add appropriate title and axis labels.
Display marks in descending order and plot again.

programming for the student marks Analysis

```python
import matplotlib.pyplot as plt

students = ['A','B','C','D','E','F','G','H','I','J']
marks = [78,85,92,67,88,95,72,81,90,76]

# Find the highest scorer
highest_idx = marks.index(max(marks))
print(f"Highest Scorer: Student {students[highest_idx]} with {marks[highest_idx]} marks")

#  Plot original marks
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.bar(students, marks, color='skyblue')
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")

# Sort in descending order and plot again
#zip pairs them, sorted sorts by marks (x[1]) in reverse, zip(*) unpacks them back
sorted_students, sorted_marks = zip(*sorted(zip(students, marks), key=lambda x: x[1], reverse=True))

plt.subplot(1, 2, 2)
plt.bar(sorted_students, sorted_marks, color='salmon')
plt.title("Marks in Descending Order")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.tight_layout()
plt.show()
```

#1. List
Definition

A List is an ordered, mutable (changeable) collection of items. It allows duplicate values and can store different data types.

Features
Ordered
Mutable
Allows duplicates
Supports indexing

```python
fruits = ["pineapple", "Grapes", "Mango"]

print(fruits)
print(fruits[0])  # Access first element
```

Important Functions
Function	Description
- append()	Add item at end
- insert()	Insert item at specific position
- remove()	Remove item
- pop()	Remove item using index
- sort()	Sort list
- reverse()	Reverse list
- len()	Count elements
- count()	Count occurrences

```python
stud = ['naga','kitty','mama']
print(stud)
print(type(stud))
stud.append("valli")
print(stud)
stud.insert(1,'abhi')
print(stud)
stud.remove('mama')
print(stud)
stud.pop(0)
print(stud)
stud.sort()
print(stud)
stud.reverse()
print(stud)
l = len(stud)
print(l)
fruit = "Banana"
fruit.count('a')
```

#3. Set
Definition

A Set is an unordered collection of unique elements.

Features
Unordered
Mutable
No duplicates allowed
Faster searching

```python
names = {"sriram", "hanuman", "venktesh"}
print(names)
```

Important Functions
Function	Description
- add()	Add element
- remove()	Remove element
- union()	Combine sets
- intersection()	Common elements
- difference()	Unique elements
- clear()	Remove all elements

```python
name1 = {'radha','shiva','krishna','dugra','laxmi'}
name2 = {'laxmi','krishna','ravana','hanuman','ashok'}
print(name1)
print(name2)
name1.add('tns')
print(name1)
name2.remove('ashok')
print(name1)
unions = name1.union(name2)
print(unions)
inter = name1.intersection(name2)
print(inter)
dif = name1.difference(name2)
print(dif)
name1.clear()
print(name1)
```

#4. Dictionary
Definition

A Dictionary stores data in key-value pairs.

Features
Mutable
Keys must be unique
Fast lookup
Stores related information together

Important Functions
Function	Description
- keys()	Returns all keys
- values()	Returns all values
- items()	Returns key-value pairs
- get()	Safely access value
- pop()	Remove key
- update()	Update dictionary

```python
student = {
    "name": "rugved",
    "age": 21,
    "course": "AIML"
}
print(student)
print(student.keys())
print(student.values())
print(student.items())
print(student.get('name'))
student.pop('name')
print(student)
student.update({'name':'hello'})
print(student)
student.update({'age':23})
print(student)
```

#Strings in Python
What is a String?

A String is a sequence of characters enclosed within single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

Features of Strings
- Strings are ordered
- Strings are immutable (cannot be modified directly)
- Supports indexing and slicing
- Can contain letters, numbers, and special characters

##String Indexing

Each character has a position called an index.

- text = "Python"
- Character	P	y	t	h	o	n
- Positive Index	0	1	2	3	4	5
- Negative Index	-6	-5	-4	-3	-2	-1

Function	Purpose
- upper()	Convert to uppercase
- lower()	Convert to lowercase
- capitalize()	First letter uppercase
- title()	Every word first letter uppercase
- strip()	Remove spaces
- replace('replaceval)	Replace text
- split()	Convert string to list
- find('str')	Find position returns index
- count('char)	Count occurrences
- startswith(value)	Check starting text returns bool
- endswith('value)	Check ending text returns bool

```python
s = 'python programming'
print(s)
print(len(s))
print(s.upper())
print(s.lower())
print(s.capitalize())
print(s.title())
print(s.replace('python','java'))
print(s.split())
print(s.find('programming'))
print(s.count('o'))
print(s.startswith("Python"))
print(s.endswith("programming"))
'''s1 = '  python  '
print(s1)

print(len(s1))
print(s1.strip())
print(s1) '''
```

```python
class Dog:
    species = "Canine"  # Class attribute

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def dog_name(self, dog, cat):
        self.dog = dog
        self.cat = cat
        print("Dog name:", self.dog)
        print("Cat name:", self.cat)


# Taking input from user
name = input("Enter dog object name: ")
age = int(input("Enter age: "))

dog_value = input("Enter dog name: ")
cat_value = input("Enter cat name: ")

# Creating object
d = Dog(name, age)

# Calling function with user input
d.dog_name(dog_value, cat_value)
```

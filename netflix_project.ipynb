```python
df = pd.read_csv("netflix_titles.csv")

print(df['type'].value_counts())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

 Which type of content is more available on Netflix

```python
plt.figure(figsize=(4,2))
sns.countplot(data=df, x="type")
plt.title("Movies and TV Shows on Netflix")
plt.show()
```

Which countries contribute the most content

```python
country = df["country"].dropna().str.split(", ").explode()

top_country = country.value_counts().head(10)

plt.figure(figsize=(12,5))
sns.barplot(x=top_country.values, y=top_country.index)
plt.title("Top 12 Countries with Netflix Content")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
```

 Which directors have the most titles on Netflix

```python
director = df["director"].dropna().str.split(", ").explode()

top_director = director.value_counts().head(12)

plt.figure(figsize=(12,5))
sns.barplot(x=top_director.values, y=top_director.index)
plt.title("Top 12 Directors on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Director")
plt.show()
```

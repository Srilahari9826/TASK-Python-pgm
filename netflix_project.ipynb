```python
df = pd.read_csv("netflix_titles.csv")

print(df['type'].value_counts())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Type of content that more avalible on Netflix

```python
df['type'].value_counts().plot(kind='bar')

plt.title("Netflix Content Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
print(df['country'].value_counts())
```

```python
df['country'].value_counts().head(12).plot(kind='bar')
plt.title("Top 12 Countries with the most content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df['country'].value_counts().head(12).plot(kind='pie')
plt.title("Top 12 Countries with the most content")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Directors with most Titles

```python

directors_counts = df['director'].value_counts()
print(directors_counts.head(12))
```

```python

plt.figure(figsize=(12, 8))
directors_counts.head(12).plot(kind='bar')
plt.title('Top 12 Directors with Most Titles on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```
